```python
import pandas as pd

data = {
    "Student_ID": [101, 102, 103, 104, 105, 106],
    "Maths": [78, 56, 90, 45, 67, 82],
    "Physics": [85, 62, 88, 50, 70, 79],
    "Chemistry": [69, 58, 91, 48, 72, 85]
}

df = pd.DataFrame(data)
df.to_csv("students.csv", index=False)

print("students.csv created successfully")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("students.csv")

df['Average'] = df[['Maths','Physics','Chemistry']].mean(axis=1)
df['Result'] = df['Average'].apply(lambda x: 'Pass' if x >= 50 else 'Fail')

print(df)

pass_count = (df['Result'] == 'Pass').sum()
fail_count = (df['Result'] == 'Fail').sum()

plt.bar(['Pass','Fail'], [pass_count, fail_count])
plt.title("Pass vs Fail Analysis")
plt.show()
subject_avg = df[['Maths','Physics','Chemistry']].mean()
print(subject_avg)


```
